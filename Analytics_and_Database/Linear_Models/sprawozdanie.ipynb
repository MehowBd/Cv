```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, classification_report
from sklearn import preprocessing
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# ZADANIE 1
* SCORES.csv


```python
df = pd.read_csv('SCORES.csv')
df
```

```python
df.head()
```

```python
df.describe()
```

```python
print(df.shape)
```

```python
fig1 = plt.figure(figsize = (20, 10))
plt.scatter(df['Hours'], df['Scores'])
plt.title('Zależność pomiędzy poświęconym czasem nauki, a otrzymanym wynikiem')
plt.xlabel('Wynik')
plt.ylabel('Czas poświęcony')
plt.grid()
plt.show()
```

```python
input_x = df['Hours']
output_y = df['Scores']

x_prac, x_test, y_prac, y_test = train_test_split(input_x, output_y, train_size = 0.3, random_state = 42)

x_prac = x_prac.to_numpy().reshape(-1, 1)
x_test = x_test.to_numpy().reshape(-1, 1)
y_prac = y_prac.to_numpy().reshape(-1, 1)
y_test = y_test.to_numpy().reshape(-1, 1)
```

Tworzenie modelu i liczenie wskaźników

```python
model = LinearRegression().fit(x_prac, y_prac)
print('Coef liczy współczynnik determinacji, określający jak dobry jest model: \n', model.coef_)
print('Intecept wylicza punkt przecięcia z osią y, dla predyktorów równych 0: \n', model.intercept_)
```

```python
predict_y = model.predict(x_test)
print('Poprawne wyniki: \n', y_test)
print('Wyniki po predykcji: \n', predict_y)
```

```python
fig2 = plt.figure(figsize = (20, 10))
plt.scatter(x_test, y_test, linewidths = 4)
plt.scatter(x_test, predict_y,linewidths = 4, color = 'g')
plt.plot(x_test, predict_y, color = 'orange')
plt.title('Zależność pomiędzy poświęconym czasem na nauke, a otrzymanym wynikiem')
plt.xlabel('Spędzony czas na naukę', fontsize = 15)
plt.ylabel('Wynik', fontsize = 15)
plt.grid()
plt.legend(['Dane testowe', 'Predykcje'], loc = 'lower right', fontsize = 15)
plt.show()
```

Średni błąd bezwzględny

```python
print(mean_absolute_error(y_test, predict_y))
```

Błąd średniokwadratowy

```python
print(mean_squared_error(y_test, predict_y))
```

Pierwiastek błędu średniokwadratowego

```python
print(np.sqrt(mean_squared_error(y_test, predict_y)))
```

* HEART.csv

```python
df = pd.read_csv('HEART.csv')
df.rename(columns = {'num       ': 'num'}, inplace = True)
df
```

```python
df.head()
```

```python
df.describe()
```

```python
print(df.shape)
```

```python
df = df.replace('?', np.NaN)
df = df.drop(columns = ['slope', 'ca', 'thal'])
df = df.dropna(axis = 0)
```

Kodowanie niezależnych zmiennych kategoralnych

```python
df = pd.get_dummies(df, columns = ['cp', 'restecg'])
df
```

```python
input_x = df.drop('num', axis = 1)
output_y = df['num']

x_prac, x_test, y_prac, y_test = train_test_split(input_x, output_y, train_size = 0.2, random_state = 42)

x_prac = x_prac.to_numpy()
x_test = x_test.to_numpy()
y_train = y_prac.to_numpy()
y_test = y_test.to_numpy()
```

Tworzenie modelu i obliczanie wskaźników

```python
model = LogisticRegression().fit(x_prac, y_prac)
print('Coef liczy współczynnik determinacji, określający jak dobry jest model: \n', model.coef_)
print('Intecept wylicza punkt przecięcia z osią y, dla predyktorów równych 0: \n', model.intercept_)
```

Testowanie modelu

```python
predict_y = model.predict(x_test)
print('Poprawne wyniki: \n', y_test)
print('Wyniki po predykcji: \n', predict_y)
```

```python
print(model.score(x_prac, y_prac))
```

```python
print(classification_report(y_test, predict_y))
```

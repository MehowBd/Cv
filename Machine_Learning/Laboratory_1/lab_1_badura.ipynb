```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
# load the famous iris data
irisRaw = load_iris()
```

```python
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
maksimum = iris["petal length (cm)"].max()
maksimum_idx = iris["petal length (cm)"].idxmax()
print("Maksimum", (maksimum, maksimum_idx))
```

```python
minimum = iris["petal length (cm)"].min()
minimum_idx = iris["petal length (cm)"].idxmin()
print("Minimum", (minimum, minimum_idx))
```
